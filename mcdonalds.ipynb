```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('./mcdonalds.csv', sep=',')
df.head()
```

```python
df.columns
```

```python
df.rename(columns={'Menu Category':'menu_category', 'Menu Items':'menu_items', \
                   'Per Serve Size':'serving_size', 'Energy (kCal)':'calories', \
                   'Protein (g)':'protein_g', 'Total fat (g)':'total_fat_g', \
                   'Sat Fat (g)':'sat_fat_g', 'Trans fat (g)':'trans_fat_g', \
                   'Cholesterols (mg)':'Cholesterols_mg', 'Total carbohydrate (g)':'total_carbohydrate_mg', \
                   'Total Sugars (g)':'total_sugars_g', 'Added Sugars (g)':'added_sugars', \
                   'Sodium (mg)':'sodium_mg'}, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,3))
sns.lmplot(x="calories",y="protein_g",data=df,hue="menu_category")
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(df.corr(), annot = True, cmap= 'coolwarm')
```

```python
#Один из способов проверить корреляцию между различными аргументами — нарисовать матрицу корреляции(тепловую карту)
```

```python
plt.figure(figsize=(16,8))
df.hist(figsize=(12,12))
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.countplot(x='menu_category', data=df, palette="Blues_d").set_title('Count of Each Category');
```
