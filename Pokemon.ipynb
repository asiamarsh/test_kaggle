```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./Pokemon.csv', sep=',')
df
```

```python
df=df.rename(columns={"#":"No", "Type 1":"Pokemon_type", 'Type 2': 'Pokemon_type2'})
df
```

```python
df.info()
```

```python
plt.figure(figsize=(3,4))
sns.displot(df['Total'])
```

```python
df['Total'].describe()
```

```python
#Распределение сил покемонов.
#минимальная сила 180
#максимальная сила 780
#средняя сила 435
# 2 моды 300 и 500
```

```python
sns.jointplot(x=df['Total'], y=df['Defense'], color='red')
```

```python
sns.kdeplot(x=df['Total'], y=df['Defense'], color='green')
```

```python
plt.figure(figsize=(14,5))
sns.countplot(x=df['Pokemon_type'])
```

```python
sns.heatmap(df.corr())
```

```python
plt.figure(figsize=(8,8))
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue="Generation")
```

```python
#Статистика Атаки и защиты покемонов
```

```python
plt.figure(figsize=(13,5))
sns.violinplot(x=df.Pokemon_type, y=df["Attack"])
```

```python
#Распределение Атак по основным типам покемонов
```

```python
plt.figure(figsize=(18,10))
sns.swarmplot(x=df.Pokemon_type, y=df["Attack"])
```

```python
plt.figure(figsize=(18,10))
sns.violinplot(x=df.Pokemon_type, y=df["Attack"])
sns.swarmplot(x=df.Pokemon_type, y=df["Attack"], color="black", alpha=0.5)
```
