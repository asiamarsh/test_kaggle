```python
#плотность распледеления по возрастам премиальных клиентов и нет
```

```python
success_logs=premium_logs.query('success == True') \
    .groupby('client', as_index=False) \
    .agg({"time":"count"}) \
    .rename(columns={"time":"succes_count"}) \
    .sort_values(["succes_count"])
success_logs
```

```python
count_succes_logs=success_logs.groupby("succes_count", as_index=False) \
    .agg({'client':'count'})
count_succes_logs
```
