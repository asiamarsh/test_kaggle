```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('./netflix_titles.csv', sep=',')
df
```

```python
df.info()
```

```python
round(df.isnull().sum()/len(df.index),2)*100
```

```python
df.dropna(axis=0, how='any', inplace=True)
```

```python
round(df.isnull().sum()/len(df.index),2)*100
```

```python
df
```

```python
df.groupby('type').count()
```

```python
df_year=df.groupby('release_year').count()
df_year=df_year.sort_values(by='release_year', ascending=False)
df_year['show_id']
```

```python
y=np.array([1940, 1945, 1950, 1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020])
```

```python
plt.figure(figsize=(15, 7))
plt.plot(df_year['show_id'], color='red', linestyle='--')
plt.xlim(1945, 2022)
plt.title('Count_films_of_year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.savefig('my_figure.png')
# plt.xticks(y)
```

```python
#мода по количеству фильмов нахидится в 2017году
```

```python
df_tv_show=df.groupby('type').get_group('TV Show')
df_tv_show
```

```python
plt.figure(figsize=(10,3))
plt.hist(df_tv_show['release_year'])
plt.xlabel("year")
plt.ylabel("count")
```

```python
df_top_country=df_tv_show.groupby('country').count().sort_values(by='show_id',ascending=False )
df_top_country=df_top_country['show_id'].head(5)
df_top_country
```

```python
country =['United States', 'United Kingdom', 'South Korea', 'Japan', 'India']
plt.figure(figsize=(5,5))
plt.pie(df_top_country, autopct='%1.1f%%', startangle=90, pctdistance=1.12)
plt.axis('equal')
plt.legend(labels=country, loc='upper right', fontsize=11, bbox_to_anchor=(1.5, 1))
```

```python
#страны где больше всего снимают TV_show
```

```python
df_movie=df.groupby('type').get_group('Movie')
df_movie
```

```python
df_movie['duration'] = pd.to_numeric(df_movie['duration'].astype(str).str.replace(r"[^\d]", "", regex=True))
df_movie
```

```python
df_movie.info()
```

```python
df_movie=df_movie.sort_values(by='duration', ascending=False)
df_movie
```

```python
plt.figure(figsize=(10,5))
plt.hist(df_movie['duration'], bins=20, color='green')
plt.xlabel('duration in minutes')
plt.ylabel('count films')
```

```python
df_movie['duration'].mean()
```

```python
#график подтверждает что средняя продолжительность фильмов равна 102мин
```

```python
df_pop_director=df_movie[['title', 'director']].groupby('director').count().sort_values(by='title', ascending=False)
df_pop_director10=df_pop_director.head(10)
df_pop_director10
```

```python
#Режиссеры с наибольшим колличеством фильмов
```

```python
df_director_year=df_movie[['title', 'director', 'release_year']].groupby(['director', 'release_year']).count().sort_values(by='title', ascending=False)
df_director_year.rename(columns={"title":"count_film_on_year"}, inplace=True)
df_director_year
```

```python
df_pop_director=df_director_year.merge(df_pop_director10, how='inner', on ="director")
df_pop_director
```
