```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./restaurants.csv', sep=',')
df
```

```python
round(df.isnull().sum()/len(df.index), 2)*100
```

```python
df.dropna(axis=0, how='any', inplace=True)
df
```

```python
df.drop(['id', 'full_address', 'zip_code', 'lat', 'lng'], axis=1, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.replace({"price_range":{'$':1, '$$':2, '$$$':3, '$$$$':4}}, inplace=True)
df
```

```python
df.describe().applymap("{:.1f}".format)
```

```python
sns.heatmap(df.corr(),annot = True)
```

```python
#оценка ресторана тесно связана с колличеством рейтингов
```

```python
top_10_restaurants= df.sort_values(by=['score', 'ratings'], ascending=False).head(10)
top_10_restaurants
```

```python
#Хороший ресторан не обязательно должен иметь высокий ценник
```

```python
df_score=df[['score', 'price_range']]
```

```python
df_score.groupby('price_range').aggregate(['max', 'mean', 'min'])
```

```python
df_score.groupby('price_range').count()
```

```python
df_expensiv=df[['name', 'score', 'ratings','price_range','category']].sort_values(by=['price_range','score','ratings'], ascending=False)[:8]
df_expensiv
```

```python
sns.lineplot(data=df_score, x='price_range', y='score')
```

```python
#лучший рейтинг у ресторанов с $-недорогим прайсом и $$-дорогим
```

```python
df_cheap=df[['name', 'score', 'ratings','price_range','category']].groupby('price_range').get_group(1)
df_cheap #вывели группу недорогих ресторанов для дальнейшего анализа
```

```python
df_cheap.sort_values(by=['ratings','score'], ascending=False).head(6)
```
